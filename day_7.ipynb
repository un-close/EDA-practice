```python
import numpy as np
x = np.array([5 , 7.2 , 9 , 6.5 , 8])         
y = np.array([6.1 , 7.1 , 8.3 , 7.6 , 6.9])  
```

```python
import matplotlib.pylab as plt
plt.scatter(x, y)
plt.show()
```

```python
#slope -> m, constant -> c
m, c = -1, 9
yh = m * x + c
yh
```

```python
plt.plot(x, yh)
plt.scatter(x, y,color='red')
plt.show()
```

```python
mse = ((y - yh) ** 2).mean()
print(mse)
```

```python
lr = 0.01
for i in range(10001):
    yh = m * x + c
    dm = -2 *((y - yh) * x).mean()
    dc = -2 *((y - yh)).mean()
    
    m, c = m - lr * dm, c - lr * dc
    mse = ((y - yh) ** 2).mean()
    if i % 100 == 0:
        print(f'{i}th iteration loss :{mse}')
    
```

```python
plt.plot(x, yh)
plt.scatter(x, y)
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('day 6/advertising.csv')
df.head()
```

```python
x = np.asarray(df.TV)
y = np.asarray(df.Sales)
print(x, y)
```

```python
plt.scatter(x, y)
```

```python

m, c = 2.1, 2
yh = m * x + c

from sklearn.metrics import mean_squared_error

print(mean_squared_error(y, yh))
plt.scatter(x, y)
plt.plot(x, yh)
```

```python
dm = -2 * ((y - yh) * x).mean()
dc = -2 * (y - yh).mean()
print(dm, dc)
```

```python
lr = 1e-6
m = m - lr * dm
c = c - lr * dc
yh = m * x + c
print(mean_squared_error(y, yh))
```

```python
for i in range(100000):
    yh = m * x + c
    dm = -2 * ((y - yh) * x).mean()
    dc = -2 * (y - yh).mean()

    m = m - lr * dm
    c = c - lr * dc
    if i % 1000 == 0:
        print(f'{i}th error:{mean_squared_error(y, yh)}')
```

```python
yh = m * x + c
plt.scatter(x, y)
plt.plot(x, yh, color='r')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = x[:, np.newaxis]
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(mean_squared_error(y, y_pred))
```
