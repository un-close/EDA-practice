```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('advertising.csv')
df.head()
```

```python
x, y = df[['TV', 'Radio']], df.Sales
x, y, b = np.asarray(x), np.asarray(y), 1
w = np.ones(x.shape[1])
print(x.shape, w.shape)
```

```python
from sklearn.metrics import mean_squared_error as mse
yh = x @ w + b
mse(y, yh)
```

```python
n = len(y)
dw = -2 * ((y - yh)/n) @ x
dw.shape
```

```python
lr = 3e-5 
limit = 10 ** 5
for i in range(limit):
    dw = -2 * np.dot((y - yh)/n, x)
    db = -2 * (y - yh).mean()
    w = w - np.dot(lr, dw)
    b = b - lr * db
    yh = np.dot(x, w) + b
print(mse(yh, y))
```
